```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000, n_features=8, random_state=1)
```

```python
from collections import Counter
count = Counter(y)
count
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import VotingClassifier
```

```python
models = list()

dectree = Pipeline([('dt',DecisionTreeClassifier())])
models.append(('dect',dectree))
randtree = Pipeline([('rt',RandomForestClassifier())])
models.append(('randt',randtree))

ensemble = VotingClassifier(estimators=models, voting='hard')

```

```python
models
```

```python
ensemble
```

```python
cv = RepeatedStratifiedKFold(n_splits=10,n_repeats=3, random_state=1)
scores = cross_val_score(ensemble,X,y,cv=cv,scoring='accuracy',n_jobs=-1)
scores
```
