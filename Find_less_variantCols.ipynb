```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv('Santander.csv',nrows=2000)
```

```python
df.head()
```

```python
X=df.drop(labels=['TARGET'], axis=1)
y=df['TARGET']
```

```python
from sklearn.model_selection import train_test_split
# separate dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    test_size=0.3,
    random_state=0)
```

```python
va = VarianceThreshold(threshold=0)
va.fit(X_train)
```

```python
sum(va.get_support())
```

```python
notin = [col for col in X_train.columns if col not in X_train.columns[va.get_support()]]
```
